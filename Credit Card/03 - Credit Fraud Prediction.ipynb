# Step 03 - Data Preprocessing

```python
df = pd.read_csv('/content/creditcard.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df['Class'].value_counts()
```

```python
# balamce the datasets
fraud = df[df['Class'] == 1]
non_fraud = df[df['Class'] == 0]
```

```python
# radom selection of simple
non_fraud_t = non_fraud.sample(n=492)
```

```python
dataset = fraud.append(non_fraud_t, ignore_index=True)
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
x = dataset.drop(columns='Class', axis=1)
y = dataset['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

# Step 04 - Create Model & training the model

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

## Evaluation the model

```python
print("Training score", accuracy_score(y_train, model.predict(x_train)))
```

```python
print('testing score', accuracy_score(y_test, model.predict(x_test)))
```
